```python
## Ejercicio 1. Realizar un algoritmo que comprueba si una cadena leída
## por teclado comienza por una subcadena introducida por teclado.

# Cadena leída por teclado y subcadena
text = input('Escriba una cadena de texto: ')
sub_text = input('Escriba la subcadena de texto: ')

# Comprobar y leer si una cadena comieza por una subcadena introducida.
## Se puede usar el operador IN para ello
def pertenece():
  if sub_text in text:
    print('pertenece a la subcadena')
  else:
    print('No pertenece')

pertenece()


```

```python
## Ejercicio 2. Algoritmo que pide una cadena y un carácter por teclado
## (valida que sea un carácter) y muestra cuántas veces aparece el carácter
## en la cadena.

## Pedir una cadena y un carácter
cadena = input('Escribe una cadena de texto: ')
caracter = input('Escribe un carácter por teclado: ')

## Validar que sea un carácter
if caracter in cadena:
  ## Mostrar cuántas veces aparece
  total = cadena.count(caracter)
  print(f'El total de ocurrencias son {total}')
else:
  print('Error, no hay ocurrencias')



```

```python
## Ejercicio 3. Suponiendo que hemos introducido una cadena
## por teclado que representa una frase (palabras separadas por espacios),
## realiza un algoritmo que cuente cuántas palabras tiene.

## Cadana por teclado
cadena = input('Introduzca una cadena por teclado (separadas por espacios): ')

## Usamos split() para comernos los espacios
div_espacios = cadena.split()

## Para contar cuantas palabras tiene tenemos que usar len()
## A diferencia de los caracteres que es con count()
total_cont = len(div_espacios)
print(total_cont)
```

```python
## Ejercicio 4. Realizar un algoritmo que lea una cadena
## por teclado y convierta las mayúsculas a minúsculas y viceversa.

cadena = input('Escribe una cadena: ')

## Convertir Mayús a minus y viceversa
## Se usa el metodo swapcase()
## Método de la clase srt ( donde tambien esta lower() y upper() )

conversor = cadena.swapcase()

print(conversor)
```
